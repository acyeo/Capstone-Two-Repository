```python
symptom_list = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia', 'Genital thrush', 'visual blurring', 'Itching', 'Irritability', 'delayed healing', 'partial paresis', 'muscle stiffness', 'Alopecia', 'Obesity', 'Diabetes Result']
groups_list = ['Age', 'Gender']
```

```python
#melt and group gender and age to find distribution of age/gender and symptoms and diabetes
melted_data = pd.melt(first_data, id_vars=['Gender', 'Age'], var_name='Symptom_Diabetes Result', value_name='Result')
grouped_data = melted_data.groupby(['Gender', 'Age', 'Symptom_Diabetes Result', 'Result']).size().reset_index(name='Count')
print(grouped_data.head())
```

```python
#visualize the distribution of symptoms and Diabetes Results by Age and Gender
plt.figure(figsize=(14, 8))
sns.barplot(x='Symptom_Diabetes Result', y='Count', hue='Result', data=grouped_data, palette='Set2')
plt.title('Distribution Symptoms and Diabetes by Age and Gender')
plt.xlabel('Symptom_Diabetes Result')
plt.ylabel('Count')
plt.legend(title='Result')
plt.xticks(rotation=90)
plt.show()
```

```python
#Look for relationship b/w average age associated with diabetes diagnoses
average_age = first_data.groupby('Diabetes Result')['Age'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Diabetes Result', y='Age', data=average_age, palette='Set2')

plt.title('Average Age by Diabetes Result')
plt.xlabel('Diabetes Result')
plt.ylabel('Average Age')

plt.show()
```

```python
#Look for relationship b/w average age and gender associated with diabetes diagnoses
# Calculate the average age for each combination of gender and diabetes result
average_age_gender = first_data.groupby(['Gender', 'Diabetes Result'])['Age'].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Diabetes Result', y='Age', hue='Gender', data=average_age_gender, palette='Set2')

# Add title and labels
plt.title('Average Age by Gender and Diabetes Result')
plt.xlabel('Diabetes Result')
plt.ylabel('Average Age')

# Show the plot
plt.show()
```

```python
#visualize at a high-level count of yes/no for symptoms and negative/positive diabetes diagnoses
categorical_columns = first_data.drop(columns=['Age', 'Gender']).columns

# Plot histograms for each categorical column
plt.figure(figsize=(20, 15))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(len(categorical_columns) // 4 + 1, 4, i)
    sns.countplot(x=column, data=first_data, palette='Set2')
    plt.title(f'Count of {column}')
    plt.xlabel('')
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#visualize at a high-level count of yes/no for symptoms that's correlatd with positive diabetes diagnoses
df_positive = first_data[first_data['Diabetes Result'] == 'Positive']


categorical_columns = df_positive.drop(columns=['Age', 'Gender', 'Diabetes Result']).columns


plt.figure(figsize=(20, 15))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(len(categorical_columns) // 4 + 1, 4, i)
    sns.countplot(x=column, data=df_positive, palette='Set2')
    plt.title(f'Count of {column} for Positive Diabetes Result')
    plt.xlabel('')
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
df_negative = first_data[first_data['Diabetes Result'] == 'Negative']

# Select categorical columns (excluding 'Age' and 'Gender')
categorical_columns = df_negative.drop(columns=['Age', 'Gender', 'Diabetes Result']).columns

# Plot histograms for each categorical column
plt.figure(figsize=(20, 15))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(len(categorical_columns) // 4 + 1, 4, i)
    sns.countplot(x=column, data=df_negative, palette='Set2')
    plt.title(f'Count of {column} for Negative Diabetes Result')
    plt.xlabel('')
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
